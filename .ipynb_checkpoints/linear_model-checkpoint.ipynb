### EDA
It isnt a bad idea to take a look at your data, even if you know your data has been cleaned.
Pandas is a great library for working with dataframes, and Scikit-learn models can use pandas dataframes.

```python
import pandas as pd
```

```python
# Import csv data amd preview
df = pd.read_csv('sample_data.csv')
df.head()
```

```python
# visualize columns for distribution shape and for any obvious correlations
pd.plotting.scatter_matrix(df, alpha=0.2);
```

```python
# Check for substantial correlation between variables
df.corr()
```

```python
# check for missing values and check datatypes
df.info()
```

### train-test split
It is generally a good idea to set some of our data aside to use for testing. This also helps prevent overfitting.

```python
from sklearn.model_selection import train_test_split
```

```python
# Split dataframe into X and y components
X = df.drop(columns=['y'])
y = df['y']
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

### Istantiate, Fit, Predict
There is so much you can do with Scikit-learn, but when it comes to supervised learning this is what it boils down to.

```python
from sklearn.linear_model import LinearRegression
```

```python
# instantiate object
linreg = LinearRegression()
```

```python
# fit model
linreg.fit(X_train, y_train)
```

```python
# predict values using model
y_train_predict = linreg.predict(X_train)
y_test_predict = linreg.predict(X_test)
```

### Coefficients and Score

```python
# inspect X1 and X2 coefficients
linreg.coef_
```

```python
# inspect y intercept
linreg.intercept_
```

```python
# inspect train R^2 value
linreg.score(X_train, y_train)
```

```python
# inspect test R^2 value
linreg.score(X_test, y_test)
```
